```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('air_data.csv', sep=";")
df
```

```python
df_count = df["Last Updated"].str.contains('2021-08-06').value_counts()
df_count
```

```python
# Supprimer la Location :
df = df.drop(columns=["Location"])
df
```

```python
# Put the "City" column Capitalized:
df['City'] = list(map(lambda x: str(x).capitalize(), df['City']))
df
```

```python
# Keep the city names before the "/" - delete what's after with regex:
df['City'] = df['City'].str.replace(r'\/.*$',r"", regex=True)
df
```

```python
# Delete row if "City" has more than one city: 
df = df[~df['City'].str.contains('-')]
df
```
